```python
# Convertir Edad a número, forzando errores a NaN
df["Edad"] = pd.to_numeric(df["Edad"], errors="coerce")

# Verificar resultado
df["Edad"].head(10)
```

```python
# Convertimos edades imposibles a NaN
df.loc[(df["Edad"] < 0) | (df["Edad"] > 120), "Edad"] = np.nan

df["Edad"].describe()
```

```python
# Calcular mediana real
mediana_edad = df["Edad"].median()

# Imputar valores faltantes con mediana
df["Edad"] = df["Edad"].fillna(mediana_edad)

# Verificar resultado
df["Edad"].describe()
```
